#### Computing the preprocessed Dataset of migrants and Austrians

```python
migrants_austrians_dis = migrants.copy()
migrants_austrians_grouped = migrants_austrians_dis.groupby(["Postal_Code","District"]).Total_Per_District.sum().reset_index()
vienna_austria2023 = vienna.merge(population2023, left_on="DISTRICT_CODE", right_on="Postal_Code")
vienna_austria2023 = vienna_austria2023.merge(migrants_austrians_grouped, left_on="DISTRICT_CODE", right_on="Postal_Code")
vienna_austria2023["Austria"]=vienna_austria2023["POP_TOTAL"]-vienna_austria2023["Total_Per_District"]
vienna_austria2023=vienna_austria2023[['Postal_Code_x','District','Austria']]
vienna_austria2023=vienna_austria2023.rename(columns={"Postal_Code_x":"Postal_Code","Austria":"Total_Per_District"})
vienna_austria2023["Country_Name"]="Austria"
for index, row in vienna_austria2023.iterrows():
    new_row = pd.DataFrame([row], columns=vienna_austria2023.columns)
    # Concatenate the new row with the existing DataFrame
    migrants_austrians_dis = pd.concat([migrants_austrians_dis, new_row], ignore_index=True)
migrants_austrians_dis=migrants_austrians_dis.sort_values("Country_Name")

# Calculate the total population for each country
total_population = migrants_austrians_dis.groupby('Country_Name')['Total_Per_District'].sum().reset_index()

# Merge the total population back into the original DataFrame
migrants_austrians = pd.merge(migrants_austrians_dis, total_population, on='Country_Name', how='left', suffixes=('_original', '_total'))

# Sort the DataFrame based on the total population
migrants_austrians = migrants_austrians.sort_values(by=['Total_Per_District_total','Postal_Code'], ascending=[False,True]).drop(columns='Total_Per_District_total')

migrants_austrians=migrants_austrians.rename(columns={"Total_Per_District_original":"Total_Per_District"})
migrants_austrians.to_csv("Datasets/migrants_austrians.csv", index=False)
migrants_austrians.info()
```

### Figure 2
Clustering countries sharing similar residence interests.

Computing z-scores to implement Map equation software

```python
%%time
# Initialize an empty dictionary to store the result
country_districts_dict = {}
countries=migrants_austrians.Country_Name.unique()
# Iterate through unique countries in the DataFrame
for country in countries:
    # Filter the DataFrame for the current country and get the list of districts
    districts = migrants_austrians[migrants_austrians['Country_Name'] == country]['Postal_Code'].tolist()
    # Add the country and its corresponding list of districts to the dictionary
    country_districts_dict[country] = districts

population=migrants_austrians['Total_Per_District'].sum()
cumulative_frac = migrants_austrians.groupby(['Country_Name']).Total_Per_District.sum()/population
edits=migrants_austrians.groupby(['Postal_Code']).Total_Per_District.sum()
dict_link_countries = {}
for country_1 in countries:
    for country_2 in countries:
        if country_1 != country_2:
            common_districts=set(country_districts_dict[country_1]).intersection(set(country_districts_dict[country_2]))
            if len(common_districts) > 0:
                for district in common_districts:

                    weight1=(migrants_austrians[(migrants_austrians['Country_Name'] == country_1) & (migrants_austrians['Postal_Code'] == district)]['Total_Per_District']).values[0]
                    weight2=(migrants_austrians[(migrants_austrians['Country_Name'] == country_2) & (migrants_austrians['Postal_Code'] == district)]['Total_Per_District']).values[0]
                    weight=weight1*weight2
                    
                    p1=cumulative_frac[country_1]
                    p2=cumulative_frac[country_2]
                    nd=edits[district]
                    m12=nd*(nd-1)*p1*p2
                    o12=nd*(nd-1)*p1*p2*((6-4*nd)*p1*p2+(nd-2)*(p1+p2)+1)
                    z12=(weight-m12)/(math.sqrt(o12))
                    # Check if country_1 and country_2 exist in the dictionary and create them if not
                    if country_1 not in dict_link_countries:
                        dict_link_countries[country_1] = {}

                    # Check if the pair has not been initialized and create it
                    if country_2 not in dict_link_countries[country_1]:
                        dict_link_countries[country_1][country_2] = {"weight": 0.0}

                    dict_link_countries[country_1][country_2]["weight"] += z12
```

```python
# Create an empty list to hold the data
data = []
# Iterate through the dictionary and extract the data
for country_1, nested_dict in dict_link_countries.items():
    for country_2, values in nested_dict.items():
        weight = values['weight']
        data.append([country_1, country_2, weight])

# Create a DataFrame from the list of data
df_zscores_aus = pd.DataFrame(data, columns=['Source', 'Target', 'Weight'])
df_zscores_aus.info()
# Save the DataFrame
df_zscores_aus.to_csv("Datasets/zscores.csv", index=False)
```

#### With Bonferroni Correction

```python
#Computing the threshold t defined in the SI.
threshold=2.82*math.sqrt(23)
# Create the 'new_column' and apply the condition
df_zscores_aus['Weights'] = 0  # Initialize the new column with zeros

for index, row in df_zscores_aus.iterrows():
    if row['Weight'] > threshold:
        df_zscores_aus.at[index, 'Weights'] = row['Weight'] - threshold
    elif row['Weight'] < -threshold:
        df_zscores_aus.at[index, 'Weights'] = row['Weight'] + threshold

# Display the updated DataFrame
df_zscores_aus.head()
```

### Figure 4
District diversity and national homophily in Vienna

The Simpson Index, which can be calculated for each district (local) and for the whole town, is a measure of ethnic concentration. It expresses the probability that two randomly selected individuals from the district/town have the same ethnicity.

```python
def citywide(df):
    total_per_country = df.groupby("Country_Name").Total_Per_District.sum()
    total = sum(total_per_country)
    citywide=0
    for i in total_per_country:
        citywide+=(i/total)**2
    citywide=1-citywide
    return citywide
```

```python
diversity=citywide(migrants_austrians)
maximum=1-21*(1/21)**2
print("Value of 0 means just one group in the city, while maximum is more diverse.")
print("Probability that two members of the population chosen at random will be of diferent subpopulation:", diversity*100)
print("The Maximum value is:", maximum*100)
```

Correlation between diversity and Fraction of Migrants

```python
vienna_merged = vienna_simpson.merge(percentage, left_on="Postal_Code", right_index=True, how="inner")
correlation, _ = stats.pearsonr(vienna_merged[percentage.name], vienna_merged["Simpson-citywide"])
n = len(vienna_merged)  # Number of data points

# Calculate the standard error of the Pearson correlation
SE_r = np.sqrt((1 - correlation**2) / (n - 2))

# Fisher transformation to compute the confidence interval
z = np.arctanh(correlation)  # Fisher transformation
SE_z = 1 / np.sqrt(n - 3)  # Standard error in z-space

# Compute the z-scores for the 95% confidence interval (1.96 corresponds to 95%)
z_lower = z - 1.96 * SE_z
z_upper = z + 1.96 * SE_z

# Inverse Fisher transformation to get back the confidence interval in r-space
r_lower = np.tanh(z_lower)
r_upper = np.tanh(z_upper)

print(f"Pearson correlation between Simpson-citywide and percentage: {correlation}")
print(f"Standard Error of correlation: {SE_r}")
print(f"95% confidence interval: [{r_lower}, {r_upper}]")
```

Dissimilarity Index for each country:
The index indicates the percentage of members of that group that should relocate to let the distributions at the districts levels match the distribution at the town level.

```python
def dissimilarity_country(group):
    total_per_district = group.groupby("Postal_Code").Total_Per_District.sum() #population per districts for one country
    total_districts = sum(total_per_district) #all population for one country
    dissimilarity=0
    denominator=2*(total_districts/Total_population)*(1-(total_districts/Total_population))
    postal=migrants_austrians.groupby("Postal_Code").Total_Per_District.sum()
    for name,i in total_per_district.items():
        dissimilarity+=(postal[name]/Total_population)*np.abs(i/postal[name]-total_districts/Total_population)
    return dissimilarity/denominator
```

```python
grouped_country = migrants_austrians.groupby("Country_Name")
# Create an empty DataFrame to store the results
result_dissimilarity = pd.DataFrame(columns=["Country_Name", "Dissimilarity"])
for Country_Name, group in grouped_country:
    result = dissimilarity_country(group)
    df_new_row = pd.DataFrame({'Country_Name': [Country_Name], 'Dissimilarity': [result]})
    result_dissimilarity = pd.concat([result_dissimilarity, df_new_row])
result_dissimilarity=result_dissimilarity.sort_values("Dissimilarity", ascending=False)
```

```python
min_height=result_dissimilarity["Dissimilarity"].min()
max_height=result_dissimilarity["Dissimilarity"].max()
lin_norm = colors.Normalize(vmin=min_height,vmax=max_height)
norm = lin_norm

cmap = plt.colormaps.get_cmap('YlOrBr')
renormalized_heights = [norm(i) for i in result_dissimilarity["Dissimilarity"]]
bar_colors = [cmap(i) for i in renormalized_heights]

fig, ax = plt.subplots(figsize=(12, 7))
# Plot the bars using the diverging colormap
bars = sns.barplot(x="Country_Name",y="Dissimilarity", data=result_dissimilarity, ax=ax, palette=bar_colors)
# Customize the plot as needed (e.g., labels, title, legend)
ax.set_xlabel('')  # This removes the default x-axis label
ax.set_ylabel("Homophily (Dissimilarity index)", fontsize=18)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
ax.tick_params(axis='both', labelsize=16)  # Adjust labelsize as needed
# Move the legend below the plot
divider = make_axes_locatable(ax)
cax = divider.append_axes('right', size='3%', pad=0.65)
colorbar = plt.colorbar(plt.cm.ScalarMappable(norm=norm, cmap=cmap),
             cax=cax, orientation='vertical', label = "Homophily")
colorbar.ax.tick_params(labelsize=18)
colorbar.set_label('Homophily', fontsize=18, labelpad=15)

lin_ticks = np.linspace(min_height, max_height, num=7)
colorbar.set_ticks(lin_ticks)
colorbar.set_ticklabels(["{:.2f}".format(tick) for tick in lin_ticks])  # Format tick labels in scientific notation

colorbar.outline.set_linewidth(0.5)  # Adjust the linewidth as needed

plt.savefig("Figures/Main/4c.pdf", bbox_inches='tight', pad_inches=0.1)
plt.show()
```
